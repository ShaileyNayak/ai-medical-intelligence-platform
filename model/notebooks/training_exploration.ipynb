# Training exploration — Chest X-Ray (Normal vs Pneumonia)

Use this notebook to inspect the dataset, sanity-check transforms, and review metrics after `train.py` / `evaluate.py`.

```python
from pathlib import Path
from collections import Counter

DATA_DIR = Path("../data")
print("Dataset root:", DATA_DIR.resolve())
for split in ["train", "test"]:
    for label in ["Normal", "Pneumonia"]:
        n = len(list((DATA_DIR / split / label).glob("*"))) if (DATA_DIR / split / label).exists() else 0
        print(f"{split}/{label}: {n}")
```

```python
import json
from IPython.display import Image, display

metrics = Path("../model/metrics/metrics_report.json")
if metrics.exists():
    print(json.dumps(json.loads(metrics.read_text()), indent=2))
cm = Path("../model/metrics/confusion_matrix.png")
if cm.exists():
    display(Image(filename=str(cm)))
```
